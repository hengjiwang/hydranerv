```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget
```

```python
from hydranerv.models.bursting.pressure.mech_sense_neuron import MechSenseNeuron
from hydranerv.models.bursting.pressure.pressure_controller import PressureController
```

```python
dt = 0.02
t_total = 200
neuron = MechSenseNeuron(dt)
pcontroller = PressureController(neuron)
time_axis = np.arange(dt, t_total, dt)
for t in time_axis:
    neuron.step(p=pcontroller.pressure())
    pcontroller.update()
neuron.disp()
pcontroller.disp()
```
