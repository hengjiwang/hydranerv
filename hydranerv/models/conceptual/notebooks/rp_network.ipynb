```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import random
from hydranerv.model.network import Network
from hydranerv.utils.utils import *
```

```python
rpnetwork = Network(type="RP", 
                  numx=30, 
                  numy=60, 
                  neuron_density=0.1, 
                  pacemaker_density=1,
                  link_maxdist=2, 
                  link_density=0.2,
                  gc=20000,
                  theta_amp=20)
rpnetwork.display()
```

```python
T = 100000
dt = 1
theta_train = np.zeros((int(T/dt)+1, 30, 60))+5
light_train = np.zeros((int(T/dt)+1, 30, 60))
i_stim_train = np.zeros((int(T/dt)+1, 30, 60))

for t in tqdm(np.arange(0, T+dt, dt)):
    rpnetwork.step(t, theta_train[int(t/dt)], light_train[int(t/dt)], i_stim_train[int(t/dt)])
```

```python
plt.figure(figsize=(10, 0.05*len(rpnetwork.neurons)))

neurons = rpnetwork.neurons

for k, neuron in enumerate(neurons.values()):
    color = randomcolor()
    for i in range(len(neuron.spike_train)):
        spike = neuron.spike_train[i]
        if spike:
            plt.vlines(i/1000, k, k+1, color)

plt.xlim(0, T/1000)
plt.ylim(0, len(rpnetwork.neurons))
plt.show()
```
