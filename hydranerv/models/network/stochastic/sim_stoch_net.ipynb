```python
import numpy as np
import matplotlib.pyplot as plt
from neuron import Neuron
from stoch_network import StochNetwork
```

## No connection

```python
N = 50
GC = 0 # 100
DENSITY = .3
TMAX = 1000
T_REF = .1
WNOISE = 0 
```

```python
# define edges
edges = []        
for i in range(N):
    for j in range(i+1, N):
        if np.random.rand() < DENSITY and (i, j) not in edges:
            edges.append((i, j))
    
# initiate network
ntwk = StochNetwork(num=N, edges=edges, gc=GC, tmax=TMAX, pacemakers=range(N), conn_type='gap_junction', wnoise=WNOISE)
```

```python
# run the simulation
ntwk.run()
```

```python
ntwk.disp(figsize=(15, 15), style='spike')
```

```python
ntwk.neurons[0].disp(figsize=(15,3))
ntwk.neurons[-1].disp(figsize=(15,3))
```
