```python
import numpy as np
import matplotlib.pyplot as plt
from rpnet.cyl_rp_network import CylRPNetwork
```

```python
ntwk = CylRPNetwork(num=100,
                      num_hyp=30,
                      num_ped=30,
                      gc=500,
                      tmax=800,
                      lambda_d=.25,
                      seed=5)
```

```python
%matplotlib widget
ntwk.disp_network(figsize=(8, 8), plot_nid=True)
```

```python
ntwk.run([35000] * 100)
```

```python
%matplotlib inline
ntwk.disp(figsize=(10, 5), style='spike')
```
