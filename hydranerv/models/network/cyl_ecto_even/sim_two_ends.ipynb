```python
import numpy as np
import matplotlib.pyplot as plt
from cyl_ecto_even_network import CylEctoEvenNetwork
```

```python
def run_sim(num_lon, prob_cb_to_rp=.2, prob_rp_to_cb=.2, prob_random=0):
    ntwk = CylEctoEvenNetwork(tmax=400,
                              num_lon=num_lon,
                              num_cir=20,
                              gc_cb=500,
                              gc_rp=500,
                              pm_cb=range(400),
                              tau_inh_cb=5,
                              tau_inh_rp=5,
                              a_inh_cb=50, # 100
                              a_inh_rp=250, # 100
                              prob_cb_to_rb=prob_cb_to_rp,
                              prob_rp_to_cb=prob_rp_to_cb,
                              prob_random=prob_random,
                              seed=114514)
    
    %matplotlib inline
    ntwk.disp_network(edge_type='cb_to_rp')
    ntwk.disp_network(edge_type='rp_to_cb')
    
    ntwk.run()
    
    ntwk.cbnet.disp()
    ntwk.rpnet.disp()
    
    plt.figure(figsize=(20,3))
    plt.plot(ntwk.cbnet.neurons[-1].v_train)
    plt.plot(ntwk.rpnet.neurons[-1].v_train)
    plt.show()
```

## intact

```python
run_sim(10, .2, .2, 1)
```

## hypostomal half

```python
run_sim(5, .2, 0, 1)
```

## peduncular half

```python
run_sim(5, 0, .2, 1)
```
