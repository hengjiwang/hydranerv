```python
import numpy as np
import matplotlib.pyplot as plt
from cyl_even_network import CylEvenNetwork
```

```python
ntwk = CylEvenNetwork(num_lon=10,
                      num_cir=20,
                      gc=700,
                      tmax=1000,
                      pacemakers=range(200),
                      wnoise=0,
                      prob_lon=1,
                      prob_cir=.2,
                      seed=1)
```

```python
# set longitudinal gradient sensitivity 
for i, neuron in enumerate(ntwk.neurons):
    (phi, z) = ntwk.locations[i]
    neuron.k_a = 8000 - (8000 - neuron.k_a) * z
    neuron.k_e = 1500 - (1500 - neuron.k_e) * z
```

```python
%matplotlib inline
ntwk.disp_network(figsize=(10, 10), show_pm=False)
```

```python
ntwk.run()
```
