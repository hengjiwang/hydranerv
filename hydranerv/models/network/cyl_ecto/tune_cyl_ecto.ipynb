```python
import numpy as np
import matplotlib.pyplot as plt
from cyl_ecto_network import CylEctoNetwork
```

```python
def plot_train(train):
    plt.figure(figsize=(20,3))
    plt.plot(train)
    plt.show()
```

```python
def run_sim(lambda_d=(.1, .05)):
    ntwk = CylEctoNetwork(tmax=400,
                          num_rp=200,
                          gc_cb=800,
                          gc_rp=500,
                          pm_cb=range(400),
                          tau_inh_cb=5,
                          tau_inh_rp=5,
                          a_inh_cb=100,
                          a_inh_rp=100,
                          lambda_d=lambda_d,
                          seed=114514)

    ntwk.run()
    ntwk.cbnet.disp()
    ntwk.rpnet.disp()

    return ntwk
```

```python
ntwk=run_sim((.1, .05))
```

```python
ntwk=run_sim((.2, .05))
```
