```python
import numpy as np
import matplotlib.pyplot as plt
from cyl_ecto_network import CylEctoNetwork
```

```python
def plot_train(train):
    plt.figure(figsize=(20,3))
    plt.plot(train)
    plt.show()
```

```python
ntwk = CylEctoNetwork(tmax=1500,
                      num_cb=100,
                      num_rp=100,
                      num_hyp=30,
                      num_ped=30,
                      gc_cb=500,
                      gc_rp=500,
                      pm_cb=range(30),
                      tau_inh_cb=5,
                      tau_inh_rp=5,
                      a_inh_cb=40,
                      a_inh_rp=40,
                      lambda_d=(.25, .25),
                      seed_cb=123,
                      seed_rp=5)

# modify pacemakers
for pm in ntwk.cbnet.pacemakers:
    neuron = ntwk.cbnet.neurons[pm]
    neuron.k_a = 7500
    neuron.k_e = 1000
```

```python
%matplotlib widget
ntwk.disp_network(edge_type='rp_to_cb', savepath='../output/rp_to_cb.png')
ntwk.disp_network(edge_type='cb_to_rp', savepath='../output/cb_to_rp.png')
ntwk.disp_network(edge_type='cb', savepath='../output/cbnet.png')
ntwk.disp_network(edge_type='rp', savepath='../output/rpnet.png')
```

```python
ntwk.run()
```
